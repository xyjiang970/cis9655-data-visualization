```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Minimum Required.

The wedges are plotted counterclockwise, by default starting from the x-axis.

```python
# make a pie chart of values. 
values = [10, 20]

plt.pie(values)

plt.show()
```

```python
# the fractional area of each wedge is given by values/sum(x)
print('The fractional area of each wedge: ', np.array(values) / sum(values))
```

## matplotlib Pie Chart 

```python
values = [5, 8, 9, 10]
colors = ['green', 'orange', 'yellow', 'grey']
labels = ['A', 'B', 'C', 'D']
explode = (0, 0, 0, 0) #specifies the fraction of the radius with which to offset each wedge.


plt.pie(values, 
        colors=colors, 
        labels=labels, 
        explode=explode,  
        autopct='%1.1f%%', #format string is used to label the wedges with their numeric value
        counterclock=True, 
        shadow=True)

plt.title('My Pie Chart')

plt.show()
```

### Pie Chart from DataFrame
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.pie.html
  
In the example below we have a DataFrame with the information about planet’s mass and radius. We pass the ‘mass’ column to the pie function to get a pie plot.

```python
# Create a DataFrame
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                    'radius': [2439.7, 6051.8, 6378.1]},
                   index=['Mercury', 'Venus', 'Earth'])
df
```

```python
#y - label or position of the column to plot. If not provided, subplots=True argument must be passed.
plot = df.plot.pie(y='mass', figsize=(5, 5))
```

```python
plot = df.plot.pie(subplots=True, figsize=(6, 3))
```

### Pie Chart from Series

```python
df['mass'].plot(kind='pie', figsize=(5, 5))
```
